```python
import pandas as pd
import requests
import json


list_of_mrt = [
'Jurong East MRT Station',
'Bukit Batok MRT Station',
'Bukit Gombak MRT Station',
'Choa Chu Kang MRT Station',
'Yew Tee MRT Station',
'Kranji MRT Station',
'Marsiling MRT Station']

df_mrt=pd.DataFrame(columns=['MRT','long','lat'])

address=[]
lats=[]
longs=[]

for i in list_of_mrt:
    req=requests.get(f'https://developers.onemap.sg/commonapi/search?searchVal={i}&returnGeom=Y&getAddrDetails=Y&pageNum=1')
    jdata = json.loads(req.text)
    if jdata['found']>=1:
        for item in jdata['results']:
            addr=item['ADDRESS']
            lat=item['LATITUDE']
            lon=item['LONGITUDE']
            address.append(addr)
            lats.append(lat)
            longs.append(lon)

row={'MRT':address,'long':longs,'lat':lats}

df=pd.DataFrame(row)
print(df)
```
